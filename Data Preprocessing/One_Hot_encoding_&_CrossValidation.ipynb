```python
import numpy as np
import pandas as pd
from seaborn import load_dataset

tips=load_dataset("tips")
tips.head()

```

```python
#ONE HOT ENCODING
from sklearn.preprocessing import OneHotEncoder
```

```python
enc=OneHotEncoder()
tra_array=enc.fit_transform(tips[["sex","smoker","day","time"]])
enc.categories_
```

```python
transformed=pd.DataFrame(tra_array.toarray(),columns=enc.get_feature_names_out())
transformed.head()
```

```python
enc_tips=pd.concat([tips,transformed],axis=1)
```

```python
enc_tips.head()
```

```python
enc_tips.drop(['sex','smoker','day','time','sex_Male','smoker_Yes','time_Lunch'],axis=1,inplace=True)
```

```python
enc_tips.head()
```

```python
enc_tips.to_csv("cleaned_tips.csv",index=True)
```

```python
#CROSS VALIDATION
```

```python
X=enc_tips.drop(['tip'],axis=1)
y=enc_tips[['tip']]
```

```python
from sklearn.model_selection import cross_validate 

from sklearn.linear_model import LinearRegression
```

```python
lin_reg=LinearRegression()
```

```python
result=cross_validate(lin_reg,X,y,cv=5,scoring='neg_mean_squared_error')
```

```python
result['test_score']
```

```python
mean_test_score=-np.mean(result['test_score'])
mean_test_score
```
